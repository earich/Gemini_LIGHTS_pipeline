```python
# This code fits a 2D Gaussian with a slopped continuum to a companion in our GPI LLP data. It was first used to fit
#    companions around MWC 789, FU Ori, and Hen 3-365 for Laws 2019 paper.

# One issue that has plegued us is that the FWHM of our objects measured with this method is much larger (~x2)
#    than the predicted FWHM value we should have given good conditions and AO performance for our observations.

import numpy as np
import scipy
import matplotlib.pyplot as plt
from astropy.modeling import models, fitting
from numpy import *
from scipy import optimize
from astropy.io import fits
import os.path
from os import path
```

```python
#From github, site when used. https://gist.github.com/andrewgiessel/6122739
#Modified from the origional to handle a slopped continuum and mask out saturated data.

def gaussian(height, center_x, center_y, width_x, width_y,xm,ym,off):
    """Returns a gaussian function with the given parameters"""
    width_x = float(width_x)
    width_y = float(width_y)
    return lambda x,y: height*exp(
        -(((center_x-x)**2./(2.*width_x**2.))+((center_y-y)**2./(2.*width_y**2.)))) + xm*x + ym*y + off

def moments(data):
    """Returns (height, x, y, width_x, width_y)
    the gaussian parameters of a 2D distribution by calculating its
    moments """
    total = data.sum()
    X, Y = indices(data.shape)
    x = (X*data).sum()/total
    y = (Y*data).sum()/total
    col = data[:, int(y)]
    width_x = sqrt(abs((arange(col.size)-y)**2*col).sum()/col.sum())
    row = data[int(x), :]
    width_y = sqrt(abs((arange(row.size)-x)**2*row).sum()/row.sum())
    height = data.max()
    off = np.nanmedian(data)
    xm, ym = 0.0, 0.0
    return height, x, y, width_x, width_y,xm,ym,off

def fitgaussian(data,mask=np.ones((40,40))):
    """Returns (height, x, y, width_x, width_y)
    the gaussian parameters of a 2D distribution found by a fit"""
    #Creates a mask to mask out saturated data. Note that in the Stokes I frame, we do not know the true saturation
    #    limit. You must go to the raw frame and based on which pixels are saturated you can create a mask.
    mask = np.copy(data)
    mask[mask<25000] = 1
    mask[mask > 2] = 0
    params = moments(data)
    errorfunction = lambda p: ravel(gaussian(*p)(*indices(data.shape))*mask - data*mask)
    p, pcov, infodict, errmsg, success = optimize.leastsq(errorfunction, params,full_output=1)
    s_sq = (errorfunction(p)**2).sum()/(len(data[0,:])*2.-len(p))
    pcov = pcov * s_sq
    error = [] 
    for i in range(len(p)):
        try:
            error.append(np.absolute(pcov[i][i])**0.5)
        except:
            error.append( 0.00 )
    perr_leastsq = np.array(error) 
    print('Best fit parameters: ',p)
    print('Error Estimates: ', perr_leastsq)
    return p,perr_leastsq
```

```python
# This function calculates basic values for the primary and companion separation based on the best fit 
#   gaussian location of the companion, the location of the primary, and the distance to the object.
#   Note that I do not include the error of the distance in these values.

def calc(sloc,comp,dist,sigma):
    pscale = (14.14,0.01) #units: milliarc/pixel #citation: https://arxiv.org/pdf/1407.2305.pdf
    r = np.sqrt((sloc[0]-comp[0])**2. + (sloc[2]-comp[2])**2.)
    dr = np.sqrt(2.*((sloc[0]-comp[0])/r)**2.*(sloc[1]**2 + comp[1]**2.) + 2.*((sloc[2]-comp[2])/r)**2.
                 **(sloc[3]**2 + comp[3]**2.))
    print('pix: ',r,dr)
    ar = (r*pscale[0]/1000.,np.sqrt((dr*pscale[0]/1000.)**2. + (r*pscale[1]/1000.)**2.))
    print('ang: ',ar[0],ar[1])
    print('AU: ', dist*ar[0], dist*ar[1])
    PA = np.arctan2(sloc[2]-comp[2],sloc[0]-comp[0])
    dtheta = np.sqrt((comp[3]/(sloc[0]-comp[0]))**2. + (sloc[3]/(sloc[0]-comp[0]))**2. 
                     + ((sloc[2]-comp[2])*sloc[1]/(sloc[0]-comp[0])**2.)**2. 
                     + ((sloc[2]-comp[2])*comp[1]/(sloc[0]-comp[0])**2.)**2.)
    theta = (sloc[2]-comp[2])/(sloc[0]-comp[0])
    
    print('PA: ', PA*180./np.pi + 90., dtheta/(1+theta**2.)*180./np.pi)
    print('FWHM :', sigma[0]*2.355,sigma[1]*2.355)
    print('\n')
```

```python
#This section of the code actually loads the data and executes the fitting functions above.

def fit_companions(name,loc):
    hdul = fits.open(name)
    images = hdul[1].data
    images = np.nan_to_num(np.copy(images[0]))
    #Size of the fitted image
    du = 10
    image = images[loc[1]-du:loc[1]+du,loc[0]-du:loc[0]+du]
    # Uncomment if you want to use the PI image. Make sure to comment out some of the lines above.
    #PI = np.sqrt(np.nan_to_num(images[1])**2. + np.nan_to_num(images[2])**2.)
    #image = PI[loc[1]-du:loc[1]+du,loc[0]-du:loc[0]+du]
    saveto = '/'.join(name.split('/')[:-1]) + '/FUOri_results.txt'

    # Creates a figure showing the best fit contour lines
    matshow(image, cmap=cm.gist_earth_r)
    # Executes the fitting routine
    params,perr_leastsq = fitgaussian(image)
    fit = gaussian(*params)

    #plots the first plot with contours
    contour(fit(*indices(image.shape)), cmap=cm.copper)
    ax = gca()
    (height, x, y, width_x, width_y,xm,ym,off) = params
    show()

    #Creates a gaussian function based on the best fit parameters.
    func = gaussian(height,x, y, width_x, width_y,xm,ym,off)
    result = np.zeros((len(image[:,0]),len(image[:,0])))
    for i in range(0,len(image[0,:])):
        for j in range(0,len(image[:,0])):
            result[j,i] = func(j,i)
    
    plt.close()

    #Creates a mask of satruated pixels.
    mask = np.copy(image)
    mask[mask<25000] = 1
    mask[mask > 2] = 0

    #plots a difference image of the actual image minus the best fit gaussian.
    plt.imshow(image*mask - result*mask, origin='lower')
    plt.plot(y,x,'x',color='w',ms = 16)
    plt.colorbar()
    plt.show()
    plt.close()
    nloc = (y + loc[0]-du,perr_leastsq[2], x + loc[1]-du,perr_leastsq[1])
    print(nloc)
    test = path.exists(saveto)
    if test == False:
        F = open(saveto, 'w')
        F.write('file, height,x, y, width_x, width_y,xm,ym,off \n')
    else:
        F = open(saveto, 'a')
    F.write(str(round(height,0)) + ',' + str(round(nloc[0],2)) + ',' + str(round(nloc[1],2)) + ',' 
            + str(round(nloc[2],2)) + ',' + str(round(nloc[3],2)) + ',' + str(round(width_x,2)) + ',' 
            + str(round(width_y,2)) + ',' + str(round(xm,1)) + ',' + str(round(ym,1)) + ',' + str(round(off,1)) + '\n')
    F.close()

    sloc = (140.0,0.1,140.0,0.1)
    #Calculating the PA,r in AU and arcsec:
    # MWC 789
    dist = 697.0
    print('FU Ori companion')
    calc(sloc,nloc,dist,((width_x + width_y)/2.,(perr_leastsq[3] + perr_leastsq[4])/2.))
    FWHM = ((width_x + width_y)/2.,(perr_leastsq[3] + perr_leastsq[4])/2.)
    return nloc, FWHM
```
